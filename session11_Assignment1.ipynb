```python
import pandas as pd
import matplotlib.pyplot as plt

#read the csv file
dat = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')

#groubby and count as per gender
fg = dat.groupby(['sex'])['sex'].count()

#set labels
labels = ['Female', 'Male']

#set gender count
count = [fg['female'], fg['male']]

#create pie chart
patches, texts = plt.pie(count, colors=colors, shadow=True, startangle=90)

#set the legwnd
plt.legend(patches, labels, loc="upper right")

#set bot axis equal
plt.axis('equal')

#show pie chart
plt.show()
```

```python
#initialize color list as per gender
color=[]

#initialize color row as per gender
#red for female and blue for male
for index, rows in dat.iterrows():
    if(rows['sex']=='female'):
        color.append('red')
    else:
        color.append('blue')

#set the figure size
plt.figure(figsize=(25,5))

#set the legend for x and y axis
plt.xlabel('Age')
plt.ylabel('Fare')

#add new column to dataframe and initialize to color
dat['color']=color

#plot the scatter plot for given data
plt.scatter(dat['age'], dat['fare'], color=dat['color'])

#show the scatter plot
plt.show()
```
